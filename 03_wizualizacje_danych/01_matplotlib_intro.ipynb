```python
# podstawowa biblioteka do wizualizacji, w sumie są teraz lepsze opcje
import matplotlib
matplotlib.__version__
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig = plt.figure() # tworzymy obiekt figure i przpisujemy do zmiennej figure
fig.show() # wyświetlamy stworzony obiekt
```

# Wykres Liniowy


```python
_ = plt.plot([0, 1], [0, 3]) # _ = przypisanie do podłogi daje to ze nie ma nr obiektu nad wykresem
```

```python
plt.plot([0, 1], [0, 3], label='line', color='green')
plt.legend()
plt.xlabel('oś X')
plt.ylabel('oś Y')
plt.title('Wykres')
```

```python
plt.plot([0, 1, 2], [0, 3, 2], label='line', color='green')
plt.legend()
plt.xlabel('oś X')
plt.ylabel('oś Y')
plt.title('Wykres')
```

```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)
plt.plot(x, x_cos)
plt.plot(x, x_sin) # można dodawać kolejne warstwy
```

# Tworzenie wielu wykresów

```python
plt.subplot(121) # 1 wiersz, 2 liczba kolumn, 1 numer wykresu
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(122)
plt.plot([0, 3, 5], [4, 3, 1], 'r--')
```

```python
# można zrobic fig recznie zeby miec kontrole nad rozmiarem itp, jak tego nie 
# zrobimy ustawi sie to automatycznie
fig = plt.figure(figsize=(14,8))
plt.subplot(131) # 1 wiersz, 2 liczba kolumn, 1 numer wykresu
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(132)
plt.plot([0, 3, 5], [4, 3, 1], 'r--')
plt.subplot(133)
plt.plot([0, 3, 5], [4, 3, 1], 'go')
```

# Wykresy słupkowe

```python
plt.bar(x=[0, 1, 2, 3], height=[1.2, 3.5, 6.9, 9.8])
```

```python
plt.bar(x=['a','b', 'c', 'd'], height=[1.2, 3.5, 6.9, 9.8])
```

```python
plt.barh(y=['a','b', 'c', 'd'], width=[1.2, 3.5, 6.9, 9.8])
```

# Wykres punktowy

```python
x1 = np.random.randn(300)
x2 = np.random.randn(300)

plt.scatter(x=x1, y=x2)
```

```python
x3 = np.random.randn(300)*50
plt.scatter(x=x1, y=x2, s=x3) # s - rozmiar kropek
```

```python
x3 = np.random.randn(300)*50
plt.scatter(x=x1, y=x2, s=x3, alpha=0.5)
```
