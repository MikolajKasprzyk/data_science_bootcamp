```python
import pandas as pd
import numpy as np
```

### <a name='a2'></a> Załadowanie danych

```python
def fetch_financial_data(company='AMZN'):
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data('META')
df.head()
```

```python
df.info()
```

### <a name='a3'></a> Zapis/Odczyt danych z pliku CSV

```python
df.to_csv('meta.csv') # zapis do pliku csv
```

```python
df_nov = df[(df.index.month == 11) & (df.index.year == 2019)]
df_nov.to_csv('meta_nov.csv')
```

```python
# read csv
new_df = pd.read_csv('meta_nov.csv')
new_df
```

```python
# zapis do pliku excel
new_df.to_excel('meta_nov.xlsx')

# odczyt excel
new_df = pd.read_excel('meta_nov.xlsx')
```

```python
new_df # kolumna z data nie jest tu indeksem
```

```python
# odczyt excel
new_df = pd.read_excel('meta_nov.xlsx', index_col=1)
new_df

# tu została nam dodatkowa kolumna która dodała się gdzieś przy zapisach i 
# odczytach tam i spowrotem, wiec ją usunę
```
